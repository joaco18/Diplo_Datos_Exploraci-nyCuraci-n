### Con los datos del archivo pasos.json hacer un histograma???(no sería un gráfico de barras) para cada provincia con la cantidad de pasos que tiene.

```python
data= pd.read_json('./input/pasos.json')
data= data[data.columns[::-1]] # aqui se invierte el orden de las columnas
data.head(8)
```

Veo cuales son las provincias que hay en el dataframe:

```python
provs = list(data['provincia'].unique())
provs
```

```python
# Tengo problemas con 'AEROPUERTO INTERNACIONAL PILOTO CIVIL NORBERTO FERNANDEZ',
# Neuquen escrito en mayúscula y minúscula los considera distintos, 
# y córdoba que esta mal escrito la considera como otra provincia mas

# Elimino 'AEROPUERTO INTERNACIONAL PILOTO CIVIL NORBERTO FERNANDEZ'
data = data[data.provincia != 'AEROPUERTO INTERNACIONAL PILOTO CIVIL NORBERTO FERNANDEZ']
# Corrijo:
data = data.replace({'NEUQUEN': 'Neuquen', 'CÃ³rdoba': 'Cordoba'})

# Vuelvo a ver la lista de provincias:
provs = list(data['provincia'].unique())
provs
```

Eliminamos los pasos duplicados

```python
cant_pasos= []

for i in provs:
    cant_pasos.append(len(data[data['provincia']==i]['name'].unique()))
```

```python
data= pd.DataFrame(zip(provs,cant_pasos))
data= data.rename(columns = {0:'provincias', 1:'cant_pasos'})
data
```

```python
plt.figure(figsize=(15,6))
sns.barplot(data=data, x='provincias', y='cant_pasos')
plt.ylabel('Cantidad de pasos')
plt.xlabel('Provincias')
plt.xticks(rotation=45)
sns.despine()
```
